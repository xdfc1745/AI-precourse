# 1주차 과제
### **음성, 언어, 이미지, 자율주행 인공지능 사례 조사**

현재 사회에서는 많은 제품과 서비스에 인공지능을 사용하고 있다. <br>
그 중 나는 내가 사용하고 있거나 흥미가 있는 제품에 대해서 조사해보았다.

## **1. 언어**
* **a. 삼성의 '빅스비 비전'**
  - 빅스비 비전은 삼성 스마트폰에 기본적으로 탑재되어 있는 기능.
  - 카메라로 비추는 언어를 즉시 번역해준다.
  - 같은 기능의 **구글의 '워드렌즈'**, **네이버의 '파파고'**가 있다.
* **b. 챗봇**
  - 사람들이 입력한 말에 알맞는 대답을 제공.
  - 예시) **세컨드마인드의 '탈라'**는 회사 내규에 대해 찾아서 답변해주는 챗봇.
  - 예시) **KLM의 '챗팅로봇'**은 인공지능 승무원으로 여행일정확인부터 체크인 알람, 항공기 발권등이 가능하다.
  



## **2. 음성**
* **a. KT의 '기가지니'**
  - 사람들이 말을 하면 그에 맞는 답변을 음성으로 제공해주는 인공지능 스피커.
  - 같은 기능의 **구글의 '알렉사'**와 **카카오의 '미니'**, **네이버의 '클로바'**, **아마존의 '에코'**가 있다.
* **b. AIA생명의 '보험상담사'**
  - 인공지능 콜센터 서비스를 제공해준다. 
  - 학습한 대화를 기반으로 고객과 대화를 나누고 계약정보를 확인, 계약을 확정하는 음성서비스 제공.
  - 비슷한 예시로 인공지능 변호사인 **'두낫페이'**가 있다.
* **c. 가비가이AB의 '돌고래 언어 해석'**
  - 돌고래 언어 데이처를 최대한 많이 모아 외국어 해석과 같은 방법으로 해석하려고함.
  - 계획대로 진행된다면 2021년에 완성될 예정.


## **3. 이미지**
* **a. 구글의 '오토드로우'**
  - 펜으로 서툴게 그림을 그려도 멋진 그림을 바꿔주는 서비스
  - 그림을 그리면 웹사이트 상단에 인공지능이 사용자 의도를 짐작해 다양한 추천그림을 제시.
  - [오토드로우](https://www.autodraw.com/)

* **b. 사물감별사**
  - **삼성의 '빅스비비전'**에는 화면에 어떠한 물건을 보여주면 그 물건에 대한 정보를 제공해준다.
  - 구글에서도 이러한 기능을 하는 **'구글렌즈'**를 현재 개발중이다.

* **c. 영화 예고편 제작**
  - IBM 인공지능 왓슨이 100여편의 공포영화를 분석, 학습하고 SF영화 '모건'의 예고편을 제작.
  - 장면을 무섭게 만드는 요소, 음악과 배우의 목소리 톤, 장면 처리나 조명의 조화를 분석.
  - 불과 24시간 만에 1분 15초짜리 예고편을 만들었다.



## **4. 자율주행**
* **a. 경주용 차**
  - 포뮬러 E의 인공지능 탑재 무인 경주용 차량 **'로보카'**.
  - 드라이버가 없이 최대 300km의 속도를 낼수 있다.

* **b.인천공항의 '에어스타**
  - 인천공항을 안내해주는 안내 로봇이다.
  - 보행자와 장애물의 움직임을 학습해 이동경로를 예측하며, 3차원 카메라 센서, 고정밀 라이더센서등을 설치하여 주행 안정성을 높임.
  - 면세점의 위치, 탑승권바코드를 인식해 탑승 게이트 위치등을 안내해준다.

* **c. SK텔레콤의 '방역로봇'**
  - 5G네트워크를 통해 서버와 실시간으로 데이터를 주고받으며 자율주행과 체온검사 등의 역할을 함.
  - 체온 검사 데이터를 서버에 보내 이를 분석하여 출입을 제한함.
